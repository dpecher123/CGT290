### numpy calculation

```python
# add (element-wise)

x = np.array([[1,2,3],[4,5,6]])
y = np.array([[1,1,1],[1,1,1]])

# 1
print(x+y)

print("-"*30)

# 2
print(np.add(x,y))
```

```python
# subtraction
x = np.array([[1,2,3],[4,5,6]])
y = np.array([[1,1,1],[1,1,1]])

# 1
print(x-y)

print("-"*30)

# 2
print(np.subtract(x,y))
```

```python
# multiply
x = np.array([[1,2,3],[4,5,6]])
y = np.array([[2,2,2],[2,2,2]])

# 1
print(x*y)

print("-"*30)

# 2
print(np.multiply(x,y))
```

```python
# divide
x = np.array([[1,2,3],[4,5,6]])
y = np.array([[2,2,2],[2,2,2]])

# 1
print(x/y)

print("-"*30)

# 2
print(np.divide(x,y))
```

```python
#3. Find the positions of max_val and min_val in a given 2D array

#(4 points)

arr = np.random.rand(50,50) * 10

def find_max_position(arr):
    
    answer1 = np.where(arr == np.amin(arr))
    coords1 = list(zip(answer1[0], answer1[1]))
    return coords1

def find_min_position(arr):
    
    answer2 = np.where(arr == np.amin(arr))
    coords2 = list(zip(answer2[0], answer2[1]))
    return coords2
    
print(find_max_position(arr))
print(find_min_position(arr))
```

```python
#4. (4 points)
# Given a randomly created matrix(100 by 100), sort the matrix in an ascending order.
# Then given a value, insert this value to the matrix

# example:
# arr = [[1,2,4],
#        [5,6,7],
#        [8,9,10]]

# val = 3

# after insertion:
# arr = [[1,2,3],
#        [5,6,7],
#        [8,9,10]]


arr = np.random.rand(100,100) * 10
val = 0.94327109

def insertion(arr, val):
    flat = arr.flatten()
    flat.sort()
    array = list(flat)
    for i in array:
        if i > val:
            array.insert(i - 1, val)
            array.remove(i)
            break
    answer = np.asarray(array)
    print(answer.reshape(50, 50))
```
